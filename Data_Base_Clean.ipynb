```python
#imports
import pandas as pd
```

```python
#read
df_2024 = pd.read_excel("CPBL_batter_2024.xlsx")
df_2025 = pd.read_excel("CPBL_batter_2025.xlsx")
df = pd.concat([df_2024, df_2025], ignore_index=True)
```

```python
#first view
df.head()
```

```python
df.info()
df.isna().sum()
```

```python
# clean
df = df.dropna() #有缺失值的欄位
df = df[df["PA"] > 100] #drop掉打席過少的noise
df = df.drop(columns = ["OPS","tOPS+"]) #drop調資訊重複的欄
df['背號'] = df['背號'].astype(str) #將被號轉為類別
```

```python
df.info()
df.describe()
```

```python
df.to_csv("base_cleaned_data.csv", index=False)
```
