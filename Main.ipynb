```python
# clean
CPBL_data = CPBL_data.dropna() #有缺失值的欄位
CPBL_data = CPBL_data[CPBL_data["PA"] > 10] #drop掉打席過少的noise
CPBL_data = CPBL_data.drop(columns = ["OPS+","tOPS+"]) #drop調資訊重複的欄
CPBL_data['背號'] = CPBL_data['背號'].astype(str) #將被號轉為類別
```
